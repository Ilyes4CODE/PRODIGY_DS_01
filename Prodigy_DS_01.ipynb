```python
import matplotlib.pyplot as pt
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('/content/Population_total_1960_to_2023.xlsx')
years = [1960, 1961, 1962, 1963, 1964, 1965, 1966, 1967, 1968, 1969,
1970, 1971, 1972, 1973, 1974, 1975, 1976, 1977, 1978, 1979,
1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989,
1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999,
2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019,
2020, 2021, 2022, 2023]

data.head()
```

```python
enter = input('Enter Your Country : ')
is_found = False
for i in range(0,len(data)):
  if enter == data.iloc[i]['Country Name']:
    liste = list(data.iloc[i])
    is_found = True
    break
```

```python
if is_found == False:
  print('Country Not Found')
else:
  liste = liste[1:]
  matrix_2d = [list(pair) for pair in zip(years, liste)]
  start = int(input('Enter Your Start Year : '))
  end = int(input('Enter Your End Year : '))
  if start < 1960:
      print("Error: Start year must be greater than or equal to 1960.")
  elif end >= 2023:
      print("Error: End year must be less than 2023.")
  elif start >= end:
      print("Error: Start year must be less than the end year.")
  else:
      print(f"Valid range: {start} to {end}")
  new_list = []
  for i in range(0,len(matrix_2d)):
    if matrix_2d[i][0] >= start and matrix_2d[i][0] <= end:
      new_list.append(matrix_2d[i])

  exacted_year , populations = [list(x) for x in zip(*new_list)]
  pt.bar(exacted_year,populations)
  pt.xticks(exacted_year)
  pt.title(enter)
  pt.xlabel('Year')
  pt.ylabel('Population')
  pt.title(f'Population Of {enter} over Years from {start} to {end}')
  pt.savefig('population_over_years.png')
  pt.show()
```
